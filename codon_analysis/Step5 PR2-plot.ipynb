### 下面的计算所有CDS的单个氨基酸的G3/(G3+C3)，A3/(A3+T3)是有误的

```python
import pandas as pd
import numpy as np
import os
from collections import defaultdict
```

```python
cusp_dir = 'RSCU'
cusp_files = [os.path.join(cusp_dir, f) for f in os.listdir(cusp_dir) if f.endswith('.txt')]
cusp_files
```

```python
def AA2codon(cusp_file):
    """获取氨基酸和密码子的对应关系.
    """
    df = pd.read_csv(cusp_files[0], comment='#', header=0, delimiter=r"\s+")[['Codon', 'AA']]
    df = df.groupby('AA')['Codon'].apply(list)
    aa2codon = df.to_dict()
    
    return aa2codon

def base3_stat(fasta):
    """计算密码子第三位的ATGC含量.
    """
    AA_stat = defaultdict(dict)
    with open(fasta) as f:
        for line in f:
            if (not line.strip().startswith('>')):
                for idx in range(0, len(line.strip()), 3):
                    codon = line[idx:idx+3]
                    try:
                        AA_stat[codon][codon[2]] += 1
                    except:
                        AA_stat[codon][codon[2]] = 1
    return AA_stat

def stat_codon2AA(aa2codon, base2codon):
    """将编码同一个氨基酸的密码子的第三位碱基计数合并到同一个氨基酸中.
    """
    stat = defaultdict(dict)

    for aa, codons in aa2codon.items():
        for codon, base_dict in base2codon.items():
            (base3, count) = list(base_dict.items())[0]
            if codon in codons:
                try:
                    stat[aa][base3] += count
                except:
                    stat[aa][base3] = count
    return stat

indir = 'filtered_data'
fastas = [os.path.join(indir, f) for f in os.listdir(indir) if f.endswith('.txt')]
fastas
```

```python
aa2codon = AA2codon(cusp_files[0])
base2codon = base3_stat(fastas[0])

stat_codon2AA(aa2codon, base2codon)
```
